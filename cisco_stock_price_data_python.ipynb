```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
# necessary imports needed for python analysis 
```

```python
df = pd.read_csv("C:\\Users\\moomo\\OneDrive\\Desktop\\cisco stock data kaggle.csv")
df.head() # print first 5 rows 
```

```python
df.info()
# displays info and types of data in table 
```

```python
df.shape[1]
# prints number of columns
```

```python
df['+/-'] = np.where(
    df['Percent change'] > 0, '+','-'
)
# added + / - change for Percent change above 0 or below 0
```

```python
df.head()
```

```python
subset = df.iloc[0:2000]

plt.figure()
plt.plot(subset['datetime'], subset['open'])
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('opening price of cisco stock over the years')
plt.show()

```

```python
subset = df.iloc[0:2000]

plt.figure()
plt.plot(subset['datetime'], subset['close'])
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('closing price of cisco stock over the years')
plt.show()
```

```python
plt.figure()

plt.plot(df['datetime'], df['open'], label='open', color='blue')
plt.plot(df['datetime'], df['close'], label='close', color='green')

plt.xlabel('date')
plt.ylabel('price')
plt.title('open vs close')
plt.legend()
plt.show()

# combined open and close graphs for easier comparison 
```
